```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
day_minusone = pd.read_csv("prices_round_2_day_-1.csv", delimiter=';')
```

```python
day_minusone.columns
```

```python
day_minusone.drop(["day", "profit_and_loss"], axis=1)
```

```python
coconuts = day_minusone[day_minusone["product"] == "COCONUTS"]
```

```python
coconuts["SMA100"] = coconuts["mid_price"].rolling(20).mean()
plt.figure(figsize=(30,18))
plt.plot(coconuts["timestamp"], coconuts["mid_price"])
plt.plot(coconuts["timestamp"], coconuts["SMA100"], color="red")
plt.show()
```

```python
pina = day_minusone[day_minusone["product"] == "PINA_COLADAS"]
```

```python
pina["SMA100"] = pina["mid_price"].rolling(20).mean()
plt.figure(figsize=(30,18))
plt.plot(pina["timestamp"], pina["mid_price"])
plt.plot(pina["timestamp"], pina["SMA100"], color="red")
plt.show()
```
